```python

# PCA Scatter
import json
import os
from pathlib import Path

os.environ.setdefault("MPLCONFIGDIR", "/tmp/matplotlib")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.legend_handler import HandlerTuple
from matplotlib.lines import Line2D

PCA_BASE_DIR = Path("/nfs/ycji_temp/code/DummyForcing/draw/pca_layers0_4_20_29")
PCA_DATA_DIR = PCA_BASE_DIR / "all_head"
PCA_SUMMARY_PATH = PCA_DATA_DIR / "summary_all_timesteps.json"
PCA_METADATA_PATH = PCA_DATA_DIR / "sample_metadata_all_timesteps.csv"
PCA_COORDS_PATH = PCA_DATA_DIR / "pca_coords_2d_all_timesteps.npy"
PCA_OUTPUT_PATH = PCA_DATA_DIR / "pca_scatter_k2_heads_4timesteps.png"
PCA_OUTPUT_CSV = PCA_DATA_DIR / "pca_scatter_k2_heads_4timesteps_with_clusters.csv"

PCA_TIMESTEPS = [1000, 937, 833, 625]
PCA_EXCLUDED_HEADS = {"layer00_head10", "layer04_head08", "layer29_head04"}
PCA_HEAD_MARKERS = ["o", "s", "D", "^", "v", "P", "X", "h", "8", "<", ">", "p"]

PCA_DYNAMIC_COLOR = "#F0A64D"
PCA_STATIC_COLOR = "#153D7A"


def pca_mix_with_white(hex_color, strength):
    hex_color = hex_color.lstrip("#")
    rgb = np.array([int(hex_color[i : i + 2], 16) for i in (0, 2, 4)], dtype=float) / 255.0
    mixed = rgb * (1.0 - strength) + strength
    return tuple(mixed.tolist())


def pca_run_kmeans_2d(points, n_clusters=2, n_init=20, max_iter=100):
    rng = np.random.default_rng(0)
    best_inertia = None
    best_labels = None
    best_centers = None

    for _ in range(n_init):
        seed_idx = rng.choice(len(points), size=n_clusters, replace=False)
        centers = points[seed_idx].copy()

        for _ in range(max_iter):
            dists = ((points[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
            labels = dists.argmin(axis=1)

            new_centers = centers.copy()
            for k in range(n_clusters):
                cluster_points = points[labels == k]
                if len(cluster_points) > 0:
                    new_centers[k] = cluster_points.mean(axis=0)

            if np.allclose(new_centers, centers):
                centers = new_centers
                break
            centers = new_centers

        final_dists = ((points[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
        final_labels = final_dists.argmin(axis=1)
        inertia = final_dists[np.arange(len(points)), final_labels].sum()

        if best_inertia is None or inertia < best_inertia:
            best_inertia = inertia
            best_labels = final_labels.copy()
            best_centers = centers.copy()

    return best_labels, best_centers


with PCA_SUMMARY_PATH.open() as f:
    pca_summary = json.load(f)

pca_metadata = pd.read_csv(PCA_METADATA_PATH)
pca_coords = np.load(PCA_COORDS_PATH)

keep_timesteps = {str(t) for t in PCA_TIMESTEPS}
pca_metadata["timestep"] = pca_metadata["timestep"].astype(str)
pca_mask = pca_metadata["timestep"].isin(keep_timesteps) & ~pca_metadata["head_label"].isin(PCA_EXCLUDED_HEADS)
pca_mask = pca_mask.to_numpy()

pca_filtered = pca_metadata.loc[pca_mask].copy().reset_index(drop=True)
pca_filtered["pc1"] = pca_coords[pca_mask, 0]
pca_filtered["pc2"] = pca_coords[pca_mask, 1]

pca_active_heads = [head for head in pca_summary["heads"] if head not in PCA_EXCLUDED_HEADS]
pca_expected = len(pca_active_heads) * len(PCA_TIMESTEPS) * 100
if len(pca_filtered) != pca_expected:
    raise ValueError(f"Expected {pca_expected} filtered samples, got {len(pca_filtered)}")

pca_labels, pca_centers_np = pca_run_kmeans_2d(pca_filtered[["pc1", "pc2"]].to_numpy(), n_clusters=2, n_init=20)
pca_filtered["cluster_raw"] = pca_labels

pca_centers = pd.DataFrame(pca_centers_np, columns=["pc1", "pc2"])
pca_centers["cluster_raw"] = pca_centers.index
pca_centers = pca_centers.sort_values("pc1").reset_index(drop=True)

pca_cluster_to_name = {
    int(pca_centers.loc[0, "cluster_raw"]): "Dynamic Head",
    int(pca_centers.loc[1, "cluster_raw"]): "Static Head",
}
pca_cluster_to_base_color = {
    int(pca_centers.loc[0, "cluster_raw"]): PCA_DYNAMIC_COLOR,
    int(pca_centers.loc[1, "cluster_raw"]): PCA_STATIC_COLOR,
}

pca_timestep_to_strength = {
    "1000": 0.00,
    "937": 0.18,
    "833": 0.36,
    "625": 0.54,
}

if len(pca_active_heads) > len(PCA_HEAD_MARKERS):
    raise ValueError(f"Need at least {len(pca_active_heads)} markers, only have {len(PCA_HEAD_MARKERS)}")
pca_head_to_marker = {head: marker for head, marker in zip(pca_active_heads, PCA_HEAD_MARKERS)}

fig, ax = plt.subplots(figsize=(13.6, 8.4), dpi=180)
fig.subplots_adjust(left=0.08, right=0.77, bottom=0.12, top=0.88)

ax.grid(True, color="#b0b0b0", alpha=0.25, linewidth=0.8)

for head in pca_active_heads:
    head_df = pca_filtered[pca_filtered["head_label"] == head]
    for timestep in PCA_TIMESTEPS:
        chunk = head_df[head_df["timestep"] == str(timestep)]
        base_colors = [pca_cluster_to_base_color[c] for c in chunk["cluster_raw"]]
        facecolors = [pca_mix_with_white(c, pca_timestep_to_strength[str(timestep)]) for c in base_colors]
        ax.scatter(
            chunk["pc1"],
            chunk["pc2"],
            s=36,
            marker=pca_head_to_marker[head],
            c=facecolors,
            alpha=0.72,
            edgecolors="white",
            linewidths=0.35,
            zorder=3,
        )

pca_evr = pca_summary["explained_variance_ratio"]
ax.set_xlabel(f"PC1 ({pca_evr[0] * 100:.1f}%)", fontsize=22)
ax.set_ylabel(f"PC2 ({pca_evr[1] * 100:.1f}%)", fontsize=22)
ax.tick_params(axis="both", labelsize=15)

pca_cluster_handles = [
    Line2D([0], [0], marker="o", linestyle="None", markersize=10, markerfacecolor=PCA_DYNAMIC_COLOR, markeredgecolor="white", markeredgewidth=0.8, label="Dynamic Head"),
    Line2D([0], [0], marker="o", linestyle="None", markersize=10, markerfacecolor=PCA_STATIC_COLOR, markeredgecolor="white", markeredgewidth=0.8, label="Static Head"),
]

legend_x = 1.02
legend_w = 0.22

cluster_legend = ax.legend(
    handles=pca_cluster_handles,
    loc="upper left",
    bbox_to_anchor=(legend_x, 0.985, legend_w, 0.01),
    mode="expand",
    frameon=True,
    fancybox=True,
    framealpha=0.92,
    fontsize=13,
    borderaxespad=0.0,
)
ax.add_artist(cluster_legend)

timestep_handles = [
    (
        Line2D([0], [0], marker="o", linestyle="None", markersize=8, markerfacecolor=pca_mix_with_white(PCA_DYNAMIC_COLOR, pca_timestep_to_strength[str(timestep)]), markeredgecolor="white", markeredgewidth=0.8),
        Line2D([0], [0], marker="o", linestyle="None", markersize=8, markerfacecolor=pca_mix_with_white(PCA_STATIC_COLOR, pca_timestep_to_strength[str(timestep)]), markeredgecolor="white", markeredgewidth=0.8),
    )
    for timestep in PCA_TIMESTEPS
]

timestep_legend = ax.legend(
    handles=timestep_handles,
    labels=[str(timestep) for timestep in PCA_TIMESTEPS],
    title="Timesteps",
    loc="center left",
    bbox_to_anchor=(legend_x, 0.54, legend_w, 0.18),
    mode="expand",
    frameon=True,
    fancybox=True,
    framealpha=0.92,
    fontsize=13,
    title_fontsize=14,
    handler_map={tuple: HandlerTuple(ndivide=None, pad=0.8)},
    borderaxespad=0.0,
)
ax.add_artist(timestep_legend)

head_handles = [
    Line2D([0], [0], marker=pca_head_to_marker[head], linestyle="None", markersize=9, markerfacecolor="#444444", markeredgecolor="white", markeredgewidth=0.8, label=head)
    for head in pca_active_heads
]

ax.legend(
    handles=head_handles,
    title="Attention Heads",
    loc="lower left",
    bbox_to_anchor=(legend_x, 0.00, legend_w, 0.36),
    mode="expand",
    frameon=True,
    fancybox=True,
    framealpha=0.92,
    fontsize=12,
    title_fontsize=14,
    borderaxespad=0.0,
)

for spine in ax.spines.values():
    spine.set_linewidth(1.1)
    spine.set_color("#222222")

fig.savefig(PCA_OUTPUT_PATH, bbox_inches="tight")

pca_filtered["cluster_name"] = pca_filtered["cluster_raw"].map(pca_cluster_to_name)
pca_filtered.to_csv(PCA_OUTPUT_CSV, index=False)

print(f"Saved figure to: {PCA_OUTPUT_PATH}")
print(f"Saved clustered metadata to: {PCA_OUTPUT_CSV}")
print(f"Filtered points: {len(pca_filtered)}")
print(f"Cluster counts: {pca_filtered['cluster_name'].value_counts().to_dict()}")
```

```python
import json
import os
from pathlib import Path

os.environ.setdefault("MPLCONFIGDIR", "/tmp/matplotlib")

import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
from matplotlib.colors import BoundaryNorm
from matplotlib.colors import ListedColormap

HEATMAP_DIR = Path("/nfs/ycji_temp/code/DummyForcing/draw/heatmap")

HEATMAP_FILES = [
    {
        "path": HEATMAP_DIR / "ar0006_layer02head10_20260330_124550_153.pt",
        "title": "Layer 2 Head 10",
        "color": "#153D7A",
        "out_name": "attention_heatmap_layer02head10_blue.png",
    },
    {
        "path": HEATMAP_DIR / "ar0006_layer14head09_20260330_124551_050.pt",
        "title": "Layer 14 Head 9",
        "color": "#F0A64D",
        "out_name": "attention_heatmap_layer14head09_orange.png",
    },
]

HEATMAP_COMBINED_OUT = HEATMAP_DIR / "attention_heatmaps_two_heads.png"
HEATMAP_STATS_OUT = HEATMAP_DIR / "attention_heatmap_stats.json"
HEATMAP_CLIP_PERCENTILE = 99.7
HEATMAP_LOW_PERCENTILE = 75.0
HEATMAP_HIGH_PERCENTILE = 94.0
HEATMAP_KEY_START = 7487
HEATMAP_SMOOTH_KERNEL = 5
HEATMAP_UNSHARP_KERNEL = 11
HEATMAP_UNSHARP_AMOUNT = 0.55


def heatmap_percentile_key(q):
    return f"p{q:g}"


def heatmap_hex_to_rgb01(hex_color):
    hex_color = hex_color.lstrip("#")
    return tuple(int(hex_color[i : i + 2], 16) / 255.0 for i in (0, 2, 4))


def heatmap_mix_with_white(hex_color, strength):
    rgb = np.array(heatmap_hex_to_rgb01(hex_color))
    mixed = rgb * (1.0 - strength) + strength
    return tuple(mixed.tolist())


def heatmap_white_to_color_cmap(hex_color):
    return ListedColormap(
        [
            "#ffffff",
            heatmap_mix_with_white(hex_color, 0.50),
            hex_color,
        ],
        name=f"white_light_strong_{hex_color.lstrip('#')}",
    )


def heatmap_compute_stats(attn):
    flat = attn.reshape(-1)
    stats = {
        "min": float(flat.min()),
        "max": float(flat.max()),
        "mean": float(flat.mean()),
        "std": float(flat.std()),
    }
    for q in sorted({HEATMAP_LOW_PERCENTILE, HEATMAP_HIGH_PERCENTILE, HEATMAP_CLIP_PERCENTILE, 90, 95, 97, 99, 99.5, 99.9, 99.99}):
        stats[heatmap_percentile_key(q)] = float(np.percentile(flat, q))
    return stats


def heatmap_smooth_and_sharpen(attn):
    tensor = torch.from_numpy(attn).float()[None, None]
    smooth = F.avg_pool2d(tensor, kernel_size=HEATMAP_SMOOTH_KERNEL, stride=1, padding=HEATMAP_SMOOTH_KERNEL // 2)
    broad = F.avg_pool2d(smooth, kernel_size=HEATMAP_UNSHARP_KERNEL, stride=1, padding=HEATMAP_UNSHARP_KERNEL // 2)
    enhanced = torch.clamp(smooth + HEATMAP_UNSHARP_AMOUNT * (smooth - broad), min=0.0)
    return enhanced[0, 0].cpu().numpy()


fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(21, 14.5), dpi=180, constrained_layout=True)
heatmap_stats_summary = {}

for ax, spec in zip(axes, HEATMAP_FILES):
    payload = torch.load(spec["path"], map_location="cpu")
    attn = payload["attn"].float().numpy()[:, HEATMAP_KEY_START:]
    attn = heatmap_smooth_and_sharpen(attn)
    stats = heatmap_compute_stats(attn)
    heatmap_stats_summary[spec["title"]] = stats

    vmax = stats[heatmap_percentile_key(HEATMAP_CLIP_PERCENTILE)]
    low = stats[heatmap_percentile_key(HEATMAP_LOW_PERCENTILE)]
    high = stats[heatmap_percentile_key(HEATMAP_HIGH_PERCENTILE)]
    if not (0.0 < low < high < vmax):
        low = max(stats["min"], vmax * 0.08)
        high = max(low * 1.5, vmax * 0.35)

    cmap = heatmap_white_to_color_cmap(spec["color"])
    norm = BoundaryNorm([0.0, low, high, vmax], cmap.N, clip=True)
    ax.imshow(
        attn,
        cmap=cmap,
        norm=norm,
        aspect="equal",
        interpolation="nearest",
        origin="upper",
    )
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel("Query Dimension", fontsize=16)
    ax.set_xlabel("Key Dimension", fontsize=16)
    for spine in ax.spines.values():
        spine.set_linewidth(1.0)
        spine.set_color("#222222")

fig.savefig(HEATMAP_COMBINED_OUT, bbox_inches="tight")
print(f"Saved combined heatmap to: {HEATMAP_COMBINED_OUT}")

for spec in HEATMAP_FILES:
    payload = torch.load(spec["path"], map_location="cpu")
    attn = payload["attn"].float().numpy()[:, HEATMAP_KEY_START:]
    attn = heatmap_smooth_and_sharpen(attn)
    stats = heatmap_stats_summary[spec["title"]]

    vmax = stats[heatmap_percentile_key(HEATMAP_CLIP_PERCENTILE)]
    low = stats[heatmap_percentile_key(HEATMAP_LOW_PERCENTILE)]
    high = stats[heatmap_percentile_key(HEATMAP_HIGH_PERCENTILE)]
    if not (0.0 < low < high < vmax):
        low = max(stats["min"], vmax * 0.08)
        high = max(low * 1.5, vmax * 0.35)

    fig_single, ax_single = plt.subplots(figsize=(21, 7.2), dpi=180, constrained_layout=True)
    cmap = heatmap_white_to_color_cmap(spec["color"])
    ax_single.imshow(
        attn,
        cmap=cmap,
        norm=BoundaryNorm([0.0, low, high, vmax], cmap.N, clip=True),
        aspect="equal",
        interpolation="nearest",
        origin="upper",
    )
    ax_single.set_xticks([])
    ax_single.set_yticks([])
    ax_single.set_ylabel("Query Dimension", fontsize=16)
    ax_single.set_xlabel("Key Dimension", fontsize=16)
    for spine in ax_single.spines.values():
        spine.set_linewidth(1.0)
        spine.set_color("#222222")

    out_path = HEATMAP_DIR / spec["out_name"]
    fig_single.savefig(out_path, bbox_inches="tight")
    print(f"Saved single heatmap to: {out_path}")
    plt.close(fig_single)

with HEATMAP_STATS_OUT.open("w") as f:
    json.dump(
        {
            "clip_percentile": HEATMAP_CLIP_PERCENTILE,
            "low_percentile": HEATMAP_LOW_PERCENTILE,
            "high_percentile": HEATMAP_HIGH_PERCENTILE,
            "smooth_kernel": HEATMAP_SMOOTH_KERNEL,
            "unsharp_kernel": HEATMAP_UNSHARP_KERNEL,
            "unsharp_amount": HEATMAP_UNSHARP_AMOUNT,
            "key_range": [HEATMAP_KEY_START, 18719],
            "stats": heatmap_stats_summary,
        },
        f,
        indent=2,
    )
print(f"Saved stats to: {HEATMAP_STATS_OUT}")
plt.close(fig)
```

## Head Type Distribution Stacked Bar

Double-panel stacked bar chart of `Static Head` and `Dynamic Head` counts across layers.


```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

BLUE_2 = "#BDD7EE"
ORANGE_2 = "#F4B183"

head_distribution_dir = Path("/ycji/code/Forcing-KV/draw/head_distribution")
head_distribution_specs = [
    {
        "path": head_distribution_dir / "forcingkv_heads_longlive_naive_sink_3_thr08.json",
        "title": "LongLive",
    },
    {
        "path": head_distribution_dir / "forcingkv_heads_selfforcing_naive_sink_3_thr08.json",
        "title": "Self-Forcing",
    },
]
head_distribution_out_path = Path("/ycji/code/Forcing-KV/draw/head_distribution/head_type_distribution_stacked_bar_pastel.png")
head_distribution_out_path.parent.mkdir(parents=True, exist_ok=True)


def build_head_count_arrays(payload):
    layers = payload["layers"]
    layer_idx = np.array([layer["layer_idx"] for layer in layers], dtype=np.int32)
    static_count = np.array([len(layer["local_head"]) for layer in layers], dtype=np.int32)
    dynamic_count = np.array([len(layer["temporal_head"]) for layer in layers], dtype=np.int32)
    return layer_idx, static_count, dynamic_count


plt.rcParams.update({
    "font.family": "STIXGeneral",
    "mathtext.fontset": "stix",
})

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(11.6, 4.0), dpi=220, sharey=True)
fig.subplots_adjust(left=0.075, right=0.99, bottom=0.19, top=0.76, wspace=0.08)

for ax, spec in zip(axes, head_distribution_specs):
    payload = json.loads(spec["path"].read_text())
    layer_idx, static_count, dynamic_count = build_head_count_arrays(payload)

    ax.bar(
        layer_idx,
        static_count,
        width=0.74,
        color=BLUE_2,
        alpha=0.96,
        edgecolor="white",
        linewidth=0.45,
        zorder=3,
    )
    ax.bar(
        layer_idx,
        dynamic_count,
        width=0.74,
        bottom=static_count,
        color=ORANGE_2,
        alpha=0.96,
        edgecolor="white",
        linewidth=0.45,
        zorder=3,
    )

    ax.set_title(spec["title"], fontsize=17, pad=7)
    ax.set_xlabel("Layer Index", fontsize=13, labelpad=8)
    ax.set_xlim(-0.7, payload["num_layers"] - 0.3)
    ax.set_ylim(0, payload["num_heads"])
    ax.set_xticks(np.arange(0, payload["num_layers"], 5))
    ax.set_xticklabels(np.arange(0, payload["num_layers"], 5), fontsize=9.5)
    ax.set_yticks(np.arange(0, payload["num_heads"] + 1, 2))
    ax.set_yticklabels(np.arange(0, payload["num_heads"] + 1, 2), fontsize=9.5)
    ax.set_axisbelow(True)
    ax.grid(True, axis="y", linestyle="-", linewidth=0.65, color="#D9DDE3", alpha=0.75)
    ax.grid(False, axis="x")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color("#333333")
    ax.spines["bottom"].set_color("#333333")
    ax.spines["left"].set_linewidth(0.9)
    ax.spines["bottom"].set_linewidth(0.9)
    ax.tick_params(axis="both", length=3.5, width=0.8, color="#333333")

axes[0].set_ylabel("# Heads", fontsize=13, labelpad=8)

legend_handles = [
    Patch(facecolor=BLUE_2, edgecolor="none", label="Static Head"),
    Patch(facecolor=ORANGE_2, edgecolor="none", label="Dynamic Head"),
]
fig.legend(handles=legend_handles, loc="upper center", bbox_to_anchor=(0.53, 0.965), ncol=2, frameon=False, fontsize=13, handlelength=1.8, columnspacing=2.4)

fig.savefig(head_distribution_out_path, bbox_inches="tight")
plt.show()
print(f"Saved to: {head_distribution_out_path}")
```

```python
from pathlib import Path
import textwrap

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

radar_out_dir = Path("/ycji/code/Forcing-KV/draw/raider")
radar_out_dir.mkdir(parents=True, exist_ok=True)

radar_metrics = [
    "temporal style",
    "subject consistency",
    "background consistency",
    "temporal flickering",
    "motion smoothness",
    "dynamic degree",
    "aesthetic quality",
    "imaging quality",
    "object class",
    "multiple objects",
    "human action",
    "color",
    "spatial relationship",
    "scene",
    "appearance style",
    "overall consistency",
]

longlive_radar_data = {
    "LongLive": {
        "temporal style": 0.2425762617588043,
        "subject consistency": 0.9692185912858178,
        "background consistency": 0.9694220609443133,
        "temporal flickering": 0.993511043624317,
        "motion smoothness": 0.9877164552097355,
        "dynamic degree": 0.4027777777777778,
        "aesthetic quality": 0.6722787414827655,
        "imaging quality": 0.6880088909513273,
        "object class": 0.9621835443037975,
        "multiple objects": 0.8827743902439025,
        "human action": 0.956,
        "color": 0.9165305855884675,
        "spatial relationship": 0.7840713951775754,
        "scene": 0.5819767441860465,
        "appearance style": 0.20537169067215286,
        "overall consistency": 0.2659267210191296,
    },
    "Dummy": {
        "color": 0.8910794672158308,
        "subject consistency": 0.9669672690957991,
        "background consistency": 0.9670119245662245,
        "temporal flickering": 0.9933529133310505,
        "motion smoothness": 0.98800046798075,
        "dynamic degree": 0.36666666666666664,
        "aesthetic quality": 0.6724177533580411,
        "imaging quality": 0.6945249882734389,
        "object class": 0.9572784810126582,
        "multiple objects": 0.8769817073170731,
        "human action": 0.952,
        "spatial relationship": 0.7699845868550398,
        "scene": 0.5835755813953488,
        "appearance style": 0.20629689429012313,
        "temporal style": 0.2424627059996128,
        "overall consistency": 0.26649049876197695,
    },
    "ForcingKV": {
        "dynamic degree": 0.45555555555555555,
        "subject consistency": 0.9667313596751127,
        "background consistency": 0.9655230145121729,
        "temporal flickering": 0.9924020374846614,
        "motion smoothness": 0.9866058668222368,
        "aesthetic quality": 0.6706959552021436,
        "imaging quality": 0.6944332382914143,
        "object class": 0.9583860759493671,
        "multiple objects": 0.8707317073170732,
        "human action": 0.956,
        "color": 0.8944485691701038,
        "spatial relationship": 0.7697681430906144,
        "scene": 0.5808139534883721,
        "appearance style": 0.2056333311899857,
        "temporal style": 0.24240313982963563,
        "overall consistency": 0.2660313543292784,
    },
}

self_forcing_radar_data = {
    "Self-Forcing": {
        "subject consistency": 0.9542445916100405,
        "background consistency": 0.9625892816587936,
        "temporal flickering": 0.9900413142394395,
        "motion smoothness": 0.9838708940884481,
        "dynamic degree": 0.6638888888888889,
        "aesthetic quality": 0.6586803439483848,
        "imaging quality": 0.695335597307799,
        "object class": 0.9405063291139241,
        "multiple objects": 0.8716463414634147,
        "human action": 0.966,
        "color": 0.8717769604133241,
        "spatial relationship": 0.8189874121806696,
        "scene": 0.5542151162790697,
        "appearance style": 0.20365161394032916,
        "temporal style": 0.2443230445086956,
        "overall consistency": 0.2684911803853127,
    },
    "Dummy": {
        "subject consistency": 0.9526664161956352,
        "background consistency": 0.9607422655682231,
        "temporal flickering": 0.9925940040737975,
        "motion smoothness": 0.9838530738838096,
        "dynamic degree": 0.6444444444444445,
        "aesthetic quality": 0.6595268405893797,
        "imaging quality": 0.6949601289766113,
        "object class": 0.9378164556962025,
        "multiple objects": 0.8858231707317074,
        "human action": 0.966,
        "color": 0.8655114384794188,
        "spatial relationship": 0.8208018897754032,
        "scene": 0.5524709302325581,
        "appearance style": 0.20438306112825863,
        "temporal style": 0.24407953545451164,
        "overall consistency": 0.26917072741254683,
    },
    "ForcingKV": {
        "subject consistency": 0.9524625817913976,
        "background consistency": 0.9601778819949128,
        "temporal flickering": 0.9909950691154581,
        "motion smoothness": 0.9828622682734822,
        "dynamic degree": 0.6916666666666667,
        "aesthetic quality": 0.6574200546869667,
        "imaging quality": 0.6943601908899621,
        "object class": 0.9390822784810127,
        "multiple objects": 0.8774390243902439,
        "human action": 0.964,
        "color": 0.8646518707629819,
        "spatial relationship": 0.8082986779437579,
        "scene": 0.5571220930232558,
        "appearance style": 0.20433176225994565,
        "temporal style": 0.2443666973412037,
        "overall consistency": 0.26843840944510633,
    },
}

radar_colors = {
    "LongLive": "#8A8F93",
    "Self-Forcing": "#8A8F93",
    "Dummy": "#8A8F93",
    "ForcingKV": "#153D7A",
}

plt.rcParams.update({
    "font.family": "STIXGeneral",
    "mathtext.fontset": "stix",
})


def scaled_radar_values(method_data, metrics, radial_floor=0.88, radial_ceiling=0.98):
    raw = np.array([[method_data[name][metric] for metric in metrics] for name in method_data], dtype=np.float64)
    mins = raw.min(axis=0)
    maxs = raw.max(axis=0)
    spans = maxs - mins
    scaled = np.empty_like(raw)

    active = spans > 1e-12
    scaled[:, active] = radial_floor + (raw[:, active] - mins[active]) / spans[active] * (radial_ceiling - radial_floor)
    scaled[:, ~active] = (radial_floor + radial_ceiling) / 2
    return {name: scaled[idx] for idx, name in enumerate(method_data)}


def wrap_metric_label(metric):
    return "\n".join(textwrap.wrap(metric.title(), width=12, break_long_words=False))


def plot_radar(method_data, title, out_path):
    angles = np.linspace(0, 2 * np.pi, len(radar_metrics), endpoint=False)
    closed_angles = np.r_[angles, angles[0]]
    scaled_data = scaled_radar_values(method_data, radar_metrics)
    plot_methods = [name for name in method_data if name != "Dummy"]

    fig, ax = plt.subplots(figsize=(5.3, 5.3), dpi=220, subplot_kw={"projection": "polar"})
    fig.subplots_adjust(left=0.10, right=0.90, bottom=0.18, top=0.86)

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_ylim(0, 1.0)
    ax.set_yticks(np.linspace(0.2, 1.0, 5))
    ax.set_yticklabels([])
    ax.set_xticks(angles)
    ax.set_xticklabels([wrap_metric_label(metric) for metric in radar_metrics], fontsize=8.8, color="#262626")
    ax.tick_params(axis="x", pad=10)

    ax.grid(color="#D0D0D0", linewidth=0.72, alpha=0.88)
    ax.spines["polar"].set_color("#222222")
    ax.spines["polar"].set_linewidth(1.1)
    ax.set_facecolor("white")

    for name in plot_methods:
        values = scaled_data[name]
        closed_values = np.r_[values, values[0]]
        color = radar_colors[name]
        ax.plot(closed_angles, closed_values, color=color, linewidth=1.45, marker="o", markersize=2.4, label=name, zorder=3)
        ax.fill(closed_angles, closed_values, color=color, alpha=0.105, zorder=2)

    ax.set_title(title, fontsize=18, pad=20, color="#222222")

    legend_handles = [
        Line2D([0], [0], color=radar_colors[name], marker="o", markersize=4, linewidth=1.6, label=name)
        for name in plot_methods
    ]
    ax.legend(
        handles=legend_handles,
        loc="lower center",
        bbox_to_anchor=(0.5, -0.18),
        ncol=2,
        frameon=False,
        fontsize=11.5,
        handlelength=1.7,
        columnspacing=1.3,
    )

    fig.savefig(out_path, dpi=300, bbox_inches="tight", facecolor="white")
    plt.show()
    return out_path


longlive_radar_out_path = plot_radar(longlive_radar_data, "LongLive", radar_out_dir / "longlive_radar.png")
self_forcing_radar_out_path = plot_radar(self_forcing_radar_data, "Self-Forcing", radar_out_dir / "self_forcing_radar.png")

print(f"Saved: {longlive_radar_out_path}")
print(f"Saved: {self_forcing_radar_out_path}")
```
