```python

# PCA Scatter
import json
import os
from pathlib import Path

os.environ.setdefault("MPLCONFIGDIR", "/tmp/matplotlib")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.legend_handler import HandlerTuple
from matplotlib.lines import Line2D

PCA_BASE_DIR = Path("/nfs/ycji_temp/code/DummyForcing/draw/pca_layers0_4_20_29")
PCA_DATA_DIR = PCA_BASE_DIR / "all_head"
PCA_SUMMARY_PATH = PCA_DATA_DIR / "summary_all_timesteps.json"
PCA_METADATA_PATH = PCA_DATA_DIR / "sample_metadata_all_timesteps.csv"
PCA_COORDS_PATH = PCA_DATA_DIR / "pca_coords_2d_all_timesteps.npy"
PCA_OUTPUT_PATH = PCA_DATA_DIR / "pca_scatter_k2_heads_4timesteps.png"
PCA_OUTPUT_CSV = PCA_DATA_DIR / "pca_scatter_k2_heads_4timesteps_with_clusters.csv"

PCA_TIMESTEPS = [1000, 937, 833, 625]
PCA_EXCLUDED_HEADS = {"layer00_head10", "layer04_head08", "layer29_head04"}
PCA_HEAD_MARKERS = ["o", "s", "D", "^", "v", "P", "X", "h", "8", "<", ">", "p"]

PCA_DYNAMIC_COLOR = "#F0A64D"
PCA_STATIC_COLOR = "#153D7A"


def pca_mix_with_white(hex_color, strength):
    hex_color = hex_color.lstrip("#")
    rgb = np.array([int(hex_color[i : i + 2], 16) for i in (0, 2, 4)], dtype=float) / 255.0
    mixed = rgb * (1.0 - strength) + strength
    return tuple(mixed.tolist())


def pca_run_kmeans_2d(points, n_clusters=2, n_init=20, max_iter=100):
    rng = np.random.default_rng(0)
    best_inertia = None
    best_labels = None
    best_centers = None

    for _ in range(n_init):
        seed_idx = rng.choice(len(points), size=n_clusters, replace=False)
        centers = points[seed_idx].copy()

        for _ in range(max_iter):
            dists = ((points[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
            labels = dists.argmin(axis=1)

            new_centers = centers.copy()
            for k in range(n_clusters):
                cluster_points = points[labels == k]
                if len(cluster_points) > 0:
                    new_centers[k] = cluster_points.mean(axis=0)

            if np.allclose(new_centers, centers):
                centers = new_centers
                break
            centers = new_centers

        final_dists = ((points[:, None, :] - centers[None, :, :]) ** 2).sum(axis=2)
        final_labels = final_dists.argmin(axis=1)
        inertia = final_dists[np.arange(len(points)), final_labels].sum()

        if best_inertia is None or inertia < best_inertia:
            best_inertia = inertia
            best_labels = final_labels.copy()
            best_centers = centers.copy()

    return best_labels, best_centers


with PCA_SUMMARY_PATH.open() as f:
    pca_summary = json.load(f)

pca_metadata = pd.read_csv(PCA_METADATA_PATH)
pca_coords = np.load(PCA_COORDS_PATH)

keep_timesteps = {str(t) for t in PCA_TIMESTEPS}
pca_metadata["timestep"] = pca_metadata["timestep"].astype(str)
pca_mask = pca_metadata["timestep"].isin(keep_timesteps) & ~pca_metadata["head_label"].isin(PCA_EXCLUDED_HEADS)
pca_mask = pca_mask.to_numpy()

pca_filtered = pca_metadata.loc[pca_mask].copy().reset_index(drop=True)
pca_filtered["pc1"] = pca_coords[pca_mask, 0]
pca_filtered["pc2"] = pca_coords[pca_mask, 1]

pca_active_heads = [head for head in pca_summary["heads"] if head not in PCA_EXCLUDED_HEADS]
pca_expected = len(pca_active_heads) * len(PCA_TIMESTEPS) * 100
if len(pca_filtered) != pca_expected:
    raise ValueError(f"Expected {pca_expected} filtered samples, got {len(pca_filtered)}")

pca_labels, pca_centers_np = pca_run_kmeans_2d(pca_filtered[["pc1", "pc2"]].to_numpy(), n_clusters=2, n_init=20)
pca_filtered["cluster_raw"] = pca_labels

pca_centers = pd.DataFrame(pca_centers_np, columns=["pc1", "pc2"])
pca_centers["cluster_raw"] = pca_centers.index
pca_centers = pca_centers.sort_values("pc1").reset_index(drop=True)

pca_cluster_to_name = {
    int(pca_centers.loc[0, "cluster_raw"]): "Dynamic Head",
    int(pca_centers.loc[1, "cluster_raw"]): "Static Head",
}
pca_cluster_to_base_color = {
    int(pca_centers.loc[0, "cluster_raw"]): PCA_DYNAMIC_COLOR,
    int(pca_centers.loc[1, "cluster_raw"]): PCA_STATIC_COLOR,
}

pca_timestep_to_strength = {
    "1000": 0.00,
    "937": 0.18,
    "833": 0.36,
    "625": 0.54,
}

if len(pca_active_heads) > len(PCA_HEAD_MARKERS):
    raise ValueError(f"Need at least {len(pca_active_heads)} markers, only have {len(PCA_HEAD_MARKERS)}")
pca_head_to_marker = {head: marker for head, marker in zip(pca_active_heads, PCA_HEAD_MARKERS)}

fig, ax = plt.subplots(figsize=(13.6, 8.4), dpi=180)
fig.subplots_adjust(left=0.08, right=0.77, bottom=0.12, top=0.88)

ax.grid(True, color="#b0b0b0", alpha=0.25, linewidth=0.8)

for head in pca_active_heads:
    head_df = pca_filtered[pca_filtered["head_label"] == head]
    for timestep in PCA_TIMESTEPS:
        chunk = head_df[head_df["timestep"] == str(timestep)]
        base_colors = [pca_cluster_to_base_color[c] for c in chunk["cluster_raw"]]
        facecolors = [pca_mix_with_white(c, pca_timestep_to_strength[str(timestep)]) for c in base_colors]
        ax.scatter(
            chunk["pc1"],
            chunk["pc2"],
            s=36,
            marker=pca_head_to_marker[head],
            c=facecolors,
            alpha=0.72,
            edgecolors="white",
            linewidths=0.35,
            zorder=3,
        )

pca_evr = pca_summary["explained_variance_ratio"]
ax.set_xlabel(f"PC1 ({pca_evr[0] * 100:.1f}%)", fontsize=22)
ax.set_ylabel(f"PC2 ({pca_evr[1] * 100:.1f}%)", fontsize=22)
ax.tick_params(axis="both", labelsize=15)

pca_cluster_handles = [
    Line2D([0], [0], marker="o", linestyle="None", markersize=10, markerfacecolor=PCA_DYNAMIC_COLOR, markeredgecolor="white", markeredgewidth=0.8, label="Dynamic Head"),
    Line2D([0], [0], marker="o", linestyle="None", markersize=10, markerfacecolor=PCA_STATIC_COLOR, markeredgecolor="white", markeredgewidth=0.8, label="Static Head"),
]

legend_x = 1.02
legend_w = 0.22

cluster_legend = ax.legend(
    handles=pca_cluster_handles,
    loc="upper left",
    bbox_to_anchor=(legend_x, 0.985, legend_w, 0.01),
    mode="expand",
    frameon=True,
    fancybox=True,
    framealpha=0.92,
    fontsize=13,
    borderaxespad=0.0,
)
ax.add_artist(cluster_legend)

timestep_handles = [
    (
        Line2D([0], [0], marker="o", linestyle="None", markersize=8, markerfacecolor=pca_mix_with_white(PCA_DYNAMIC_COLOR, pca_timestep_to_strength[str(timestep)]), markeredgecolor="white", markeredgewidth=0.8),
        Line2D([0], [0], marker="o", linestyle="None", markersize=8, markerfacecolor=pca_mix_with_white(PCA_STATIC_COLOR, pca_timestep_to_strength[str(timestep)]), markeredgecolor="white", markeredgewidth=0.8),
    )
    for timestep in PCA_TIMESTEPS
]

timestep_legend = ax.legend(
    handles=timestep_handles,
    labels=[str(timestep) for timestep in PCA_TIMESTEPS],
    title="Timesteps",
    loc="center left",
    bbox_to_anchor=(legend_x, 0.54, legend_w, 0.18),
    mode="expand",
    frameon=True,
    fancybox=True,
    framealpha=0.92,
    fontsize=13,
    title_fontsize=14,
    handler_map={tuple: HandlerTuple(ndivide=None, pad=0.8)},
    borderaxespad=0.0,
)
ax.add_artist(timestep_legend)

head_handles = [
    Line2D([0], [0], marker=pca_head_to_marker[head], linestyle="None", markersize=9, markerfacecolor="#444444", markeredgecolor="white", markeredgewidth=0.8, label=head)
    for head in pca_active_heads
]

ax.legend(
    handles=head_handles,
    title="Attention Heads",
    loc="lower left",
    bbox_to_anchor=(legend_x, 0.00, legend_w, 0.36),
    mode="expand",
    frameon=True,
    fancybox=True,
    framealpha=0.92,
    fontsize=12,
    title_fontsize=14,
    borderaxespad=0.0,
)

for spine in ax.spines.values():
    spine.set_linewidth(1.1)
    spine.set_color("#222222")

fig.savefig(PCA_OUTPUT_PATH, bbox_inches="tight")

pca_filtered["cluster_name"] = pca_filtered["cluster_raw"].map(pca_cluster_to_name)
pca_filtered.to_csv(PCA_OUTPUT_CSV, index=False)

print(f"Saved figure to: {PCA_OUTPUT_PATH}")
print(f"Saved clustered metadata to: {PCA_OUTPUT_CSV}")
print(f"Filtered points: {len(pca_filtered)}")
print(f"Cluster counts: {pca_filtered['cluster_name'].value_counts().to_dict()}")
```

```python
import json
import os
from pathlib import Path

os.environ.setdefault("MPLCONFIGDIR", "/tmp/matplotlib")

import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
from matplotlib.colors import BoundaryNorm
from matplotlib.colors import ListedColormap

HEATMAP_DIR = Path("/nfs/ycji_temp/code/DummyForcing/draw/heatmap")

HEATMAP_FILES = [
    {
        "path": HEATMAP_DIR / "ar0006_layer02head10_20260330_124550_153.pt",
        "title": "Layer 2 Head 10",
        "color": "#153D7A",
        "out_name": "attention_heatmap_layer02head10_blue.png",
    },
    {
        "path": HEATMAP_DIR / "ar0006_layer14head09_20260330_124551_050.pt",
        "title": "Layer 14 Head 9",
        "color": "#F0A64D",
        "out_name": "attention_heatmap_layer14head09_orange.png",
    },
]

HEATMAP_COMBINED_OUT = HEATMAP_DIR / "attention_heatmaps_two_heads.png"
HEATMAP_STATS_OUT = HEATMAP_DIR / "attention_heatmap_stats.json"
HEATMAP_CLIP_PERCENTILE = 99.7
HEATMAP_LOW_PERCENTILE = 75.0
HEATMAP_HIGH_PERCENTILE = 94.0
HEATMAP_KEY_START = 7487
HEATMAP_SMOOTH_KERNEL = 5
HEATMAP_UNSHARP_KERNEL = 11
HEATMAP_UNSHARP_AMOUNT = 0.55


def heatmap_percentile_key(q):
    return f"p{q:g}"


def heatmap_hex_to_rgb01(hex_color):
    hex_color = hex_color.lstrip("#")
    return tuple(int(hex_color[i : i + 2], 16) / 255.0 for i in (0, 2, 4))


def heatmap_mix_with_white(hex_color, strength):
    rgb = np.array(heatmap_hex_to_rgb01(hex_color))
    mixed = rgb * (1.0 - strength) + strength
    return tuple(mixed.tolist())


def heatmap_white_to_color_cmap(hex_color):
    return ListedColormap(
        [
            "#ffffff",
            heatmap_mix_with_white(hex_color, 0.50),
            hex_color,
        ],
        name=f"white_light_strong_{hex_color.lstrip('#')}",
    )


def heatmap_compute_stats(attn):
    flat = attn.reshape(-1)
    stats = {
        "min": float(flat.min()),
        "max": float(flat.max()),
        "mean": float(flat.mean()),
        "std": float(flat.std()),
    }
    for q in sorted({HEATMAP_LOW_PERCENTILE, HEATMAP_HIGH_PERCENTILE, HEATMAP_CLIP_PERCENTILE, 90, 95, 97, 99, 99.5, 99.9, 99.99}):
        stats[heatmap_percentile_key(q)] = float(np.percentile(flat, q))
    return stats


def heatmap_smooth_and_sharpen(attn):
    tensor = torch.from_numpy(attn).float()[None, None]
    smooth = F.avg_pool2d(tensor, kernel_size=HEATMAP_SMOOTH_KERNEL, stride=1, padding=HEATMAP_SMOOTH_KERNEL // 2)
    broad = F.avg_pool2d(smooth, kernel_size=HEATMAP_UNSHARP_KERNEL, stride=1, padding=HEATMAP_UNSHARP_KERNEL // 2)
    enhanced = torch.clamp(smooth + HEATMAP_UNSHARP_AMOUNT * (smooth - broad), min=0.0)
    return enhanced[0, 0].cpu().numpy()


fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(21, 14.5), dpi=180, constrained_layout=True)
heatmap_stats_summary = {}

for ax, spec in zip(axes, HEATMAP_FILES):
    payload = torch.load(spec["path"], map_location="cpu")
    attn = payload["attn"].float().numpy()[:, HEATMAP_KEY_START:]
    attn = heatmap_smooth_and_sharpen(attn)
    stats = heatmap_compute_stats(attn)
    heatmap_stats_summary[spec["title"]] = stats

    vmax = stats[heatmap_percentile_key(HEATMAP_CLIP_PERCENTILE)]
    low = stats[heatmap_percentile_key(HEATMAP_LOW_PERCENTILE)]
    high = stats[heatmap_percentile_key(HEATMAP_HIGH_PERCENTILE)]
    if not (0.0 < low < high < vmax):
        low = max(stats["min"], vmax * 0.08)
        high = max(low * 1.5, vmax * 0.35)

    cmap = heatmap_white_to_color_cmap(spec["color"])
    norm = BoundaryNorm([0.0, low, high, vmax], cmap.N, clip=True)
    ax.imshow(
        attn,
        cmap=cmap,
        norm=norm,
        aspect="equal",
        interpolation="nearest",
        origin="upper",
    )
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel("Query Dimension", fontsize=16)
    ax.set_xlabel("Key Dimension", fontsize=16)
    for spine in ax.spines.values():
        spine.set_linewidth(1.0)
        spine.set_color("#222222")

fig.savefig(HEATMAP_COMBINED_OUT, bbox_inches="tight")
print(f"Saved combined heatmap to: {HEATMAP_COMBINED_OUT}")

for spec in HEATMAP_FILES:
    payload = torch.load(spec["path"], map_location="cpu")
    attn = payload["attn"].float().numpy()[:, HEATMAP_KEY_START:]
    attn = heatmap_smooth_and_sharpen(attn)
    stats = heatmap_stats_summary[spec["title"]]

    vmax = stats[heatmap_percentile_key(HEATMAP_CLIP_PERCENTILE)]
    low = stats[heatmap_percentile_key(HEATMAP_LOW_PERCENTILE)]
    high = stats[heatmap_percentile_key(HEATMAP_HIGH_PERCENTILE)]
    if not (0.0 < low < high < vmax):
        low = max(stats["min"], vmax * 0.08)
        high = max(low * 1.5, vmax * 0.35)

    fig_single, ax_single = plt.subplots(figsize=(21, 7.2), dpi=180, constrained_layout=True)
    cmap = heatmap_white_to_color_cmap(spec["color"])
    ax_single.imshow(
        attn,
        cmap=cmap,
        norm=BoundaryNorm([0.0, low, high, vmax], cmap.N, clip=True),
        aspect="equal",
        interpolation="nearest",
        origin="upper",
    )
    ax_single.set_xticks([])
    ax_single.set_yticks([])
    ax_single.set_ylabel("Query Dimension", fontsize=16)
    ax_single.set_xlabel("Key Dimension", fontsize=16)
    for spine in ax_single.spines.values():
        spine.set_linewidth(1.0)
        spine.set_color("#222222")

    out_path = HEATMAP_DIR / spec["out_name"]
    fig_single.savefig(out_path, bbox_inches="tight")
    print(f"Saved single heatmap to: {out_path}")
    plt.close(fig_single)

with HEATMAP_STATS_OUT.open("w") as f:
    json.dump(
        {
            "clip_percentile": HEATMAP_CLIP_PERCENTILE,
            "low_percentile": HEATMAP_LOW_PERCENTILE,
            "high_percentile": HEATMAP_HIGH_PERCENTILE,
            "smooth_kernel": HEATMAP_SMOOTH_KERNEL,
            "unsharp_kernel": HEATMAP_UNSHARP_KERNEL,
            "unsharp_amount": HEATMAP_UNSHARP_AMOUNT,
            "key_range": [HEATMAP_KEY_START, 18719],
            "stats": heatmap_stats_summary,
        },
        f,
        indent=2,
    )
print(f"Saved stats to: {HEATMAP_STATS_OUT}")
plt.close(fig)
```
